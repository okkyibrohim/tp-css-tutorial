# **OpenRouter**

With the ease of use and accessibility of large language models (LLMs) through APIs, there has been a surge in the number of LLM providers. Each of these providers offers different models, pricing, and features. Although this variety is beneficial for users, it can also be overwhelming to navigate and manage multiple APIs. This is where [OpenRouter](https://openrouter.ai/) comes in.

OpenRouter is a platform that provides a unified interface to access multiple LLM providers. It acts as a router, directing API calls to the appropriate provider based on user preferences and requirements. This allows users to easily switch between different LLMs without having to manage multiple APIs and credentials.

Because of aggregator platforms like OpenRouter, users are able to use a single entry point with a single payment method to access multiple LLM providers. Other than simplify the process, this enable users to purchase LLM services in one place rather than having to sign up for multiple accounts across different providers.

## Platform Overview

### Account Creation and Registration

To use OpenRouter, you need to create an account on their platform.
Go to [OpenRouter page](https://openrouter.ai) and sign up for an account.

![OpenRouter Signup](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-entry.png)

Then input the credentials to register, and wait for the registration process to complete.

![OpenRouter CredentialInput](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-input.png)

After registration, you can log in to your OpenRouter account and access the dashboard where you can manage your API keys, view usage, and configure your routing preferences.

### Model Catalog

As mentioned before, OpenRouter provides access to multiple LLM providers through a single interface. To see the available models, you can navigate to the "Models" section in the dashboard. Here, you will find a catalog of models from various providers, along with their specifications and pricing information

![OpenRouter Models Entrypoint](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-model-entry.png)

In the model catalog, you can browse through the different models, filter them based on your requirements, and view detailed information about each model, including its capabilities, pricing, and usage limits.

For example, if you want to only use free models, you can apply the "Free" filter by setting the *Prompt pricing* filter to "Free". This will show you all the models that are available for free.

![OpenRouter Model Catalog](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-model.png)

### Managing API Keys

To be able to use the models through OpenRouter, you need to manage your API keys. In the dashboard, you can navigate to the "API Keys" section where you can create and manage your API keys for different providers.

From the main page, click on the "Settings" tab to go to the account dashboard, then click on the "API Keys" tab in the left sidebar to manage your API keys.

![OpenRouter Dashboard](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-login.png)

In the API Keys page, you can add your API keys by clicking on the "Create" button and filling in the required information.

![OpenRouter API Keys](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-key-entry.png)

Fill in the name, credit limit, limit reset period, and the expiration date for the API key, then click "Create" to save the API key.

![OpenRouter API Key Creation](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-key-create.png)

After that it will show you the API key that you can copy and use in your applications to access the models through OpenRouter.

Makes sure to copy the API key and store it securely, as you won't be able to see it again after you navigate away from the page. If you lose the API key, you will need to create a new one.

![OpenRouter API Key Creation New](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-key-copy.png)

You can create multiple API keys for different purposes (e.g different projects, different providers, etc.) and manage them all in one place through the dashboard.

### Credit Management

For usage of paid models, you need to manage your credits.

To go to the credit dashboard, click on the "Credit" tab in the left sidebar from the account dashboard. Here you can view your current credit balance, add credits to your account, and view your credit usage history.

![OpenRouter Credit](https://raw.githubusercontent.com/okkyibrohim/tp-css-tutorial/main/figure/or-credit.png)

Add credits to your account by clicking on the "Add Credits" button and following the instructions to complete the payment process. You can also view your credit usage history to see how much credit you have used and when it was used.

## OpenAI Compatible API

As we know it OpenAI is the one popularize LLM as API, therefore as a lot of users are familiar with OpenAI API, many LLM providers choose to make their API compatible with OpenAI API. This means that users can use the same code and libraries they use for OpenAI API to access other LLM providers. This compatibility makes it easier for users to switch between different LLM providers without having to learn new APIs or change their existing codebase.

OpenRouter takes advantage of this OpenAI compatibility by allowing users to route their API calls to different LLM providers that support the OpenAI API. This means that users can easily switch between different LLM providers without having to change their code or learn new APIs, as long as the providers they want to use are compatible with the OpenAI API.

By this extension, we can use the OpenAI API client libraries to access multiple LLM providers [through OpenRouter](https://openrouter.ai/docs/quickstart#using-the-openai-sdk), making it easier for users to manage and utilize different LLM services without the hassle of dealing with multiple APIs and credentials.

```python
# install openai library
!pip install openai
```

```python
from openai import OpenAI

model = "liquid/lfm-2.5-1.2b-instruct:free"

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="<OPENROUTER_API_KEY>",
)

result = client.chat.completions.parse(
    model=model,
    messages=[
        {
            "role": "user",
            "content": "What is the capital of France?",
        }
    ],
)

print(result.choices[0].message.content)
```

Thats it, it as easy as that to use multiple LLM providers through a single interface with OpenRouter, and the OpenAI API compatibility makes it even easier to switch between different providers without having to change your code or learn new APIs.
