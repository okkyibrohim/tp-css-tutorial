# **Regular Expression**

Regular Expression (regex) is a powerful tool for searching and manipulating strings based on specific patterns. It allows you to define complex search patterns using a combination of characters, metacharacters, and quantifiers. In Python, the [`re`](https://docs.python.org/3/library/re.html) module provides functions for working with regular expressions.

## When to Use Regular Expressions

For most of the cases with text processing, we can use the built-in string methods or libraries like spaCy to manipulate and process the text data. However, there are certain situations where regular expressions can be particularly useful that are not easily handled by other methods. Here are some examples of when to use regular expressions:
- **Pattern Matching**: When you need to search for specific patterns in a string, such as email addresses, phone numbers, or URLs, regular expressions can provide a concise and efficient way to do so.
- **Data Validation**: Regular expressions can be used to validate input data, such as checking if a string is in the correct format for an email address or a date.
- **Data Extraction**: When you need to extract specific information from a string, such as extracting all the hashtags from a tweet or all the words that start with a certain letter, regular expressions can help you achieve this efficiently.
- **Complex String Manipulation**: For more complex string manipulation tasks that involve multiple conditions or patterns, regular expressions can provide a powerful and flexible solution.

Overall, regular expressions are a valuable tool for text processing and can be used in conjunction with other methods to achieve more efficient and effective results.

## Regular Expression Syntax

Even for experienced programmers, regular expressions can be difficult to read and understand due to their compact and symbolic nature. However, understanding the basic syntax of regular expressions is essential for using them effectively. Here are some of the key components of regular expression syntax:
- **Literals**: These are the basic characters that you want to match in a string. For example, the regex `cat` will match the string "cat" in a larger text.
- **Metacharacters**: These are special characters that have a specific meaning in regular expressions. For example, the dot `.` matches any character except a newline, and the asterisk `*` matches zero or more occurrences of the preceding character. Or something like `\d` matches any digit, and `\w` matches any word character (alphanumeric or underscore).
- **Character Classes**: These allow you to define a set of characters to match. For example, `[aeiou]` will match any vowel in a string.
- **Quantifiers**: These specify how many times a character or group should be matched. For example, `\d{3}` will match exactly three digits in a row.
- **Anchors**: These are used to specify the position of the match in the string. For example, `^` matches the start of a string, and `$` matches the end of a string.


For example the patter 'key-[aiueo]{3}-\d{2}' will match a string that starts with "key-" followed by three vowels and two digits, such as "key-aei-12" or "key-iou-34".

Using those components, you can create complex regular expressions to match a wide variety of patterns in text data.

## Pattern Matching

Pattern Matching is the task of searching for specific patterns in a string.

The `re` module provides several functions for pattern matching, including `re.search()` and `re.match()`.

```python
import re

# excerpt from https://www.bbc.com/news/articles/czx499e6y9wo
news = """
A woodland classroom has been built at a popular nature reserve.
Northumberland Wildlife Trust has opened the facility at its site in Hauxley, near Morpeth, to provide an outdoor learning space for children and adults.
It replaces a playground, which the trust said had reached the end of its life.
The charity's Druridge Bay landscapes manager Alex Lister said wildlife-rich environments helped people's mental and physical health.
"People who have the opportunity to experience nature on their doorstep are more active and mentally resilient," he said.
The classroom is made using natural wood, including some from the trust's other reserves, and the organisation said it hoped it would complement the indoor learning space.
Baby and toddler groups, Scouts, Brownies, schools and creative groups are expected to make use of the space.
The classroom was funded through a £8,100 grant from the Nadara Sisters North Steads Wind Farm Community Benefit Fund.
"The new classroom will help more of the Hauxley community and beyond access nature in a new and different way," Lister added. 
"""
```

Example of pattern matching is we want to find the first occurrence of a pattern in a string. We can use the `re.search()` function for this purpose.

For example, if we want to find the first occurence of the word 'Nature' in the news.

```python
pattern = "nature"
match = re.search(pattern, news)

if match:
    print(f"Found '{pattern}' at position {match.start()}-{match.end()}")
else:
    print(f"'{pattern}' not found in the text.")
```

In python, 'r' before the string indicates that it is a raw string, which means that backslashes are treated as literal characters and not as escape characters. This is important when writing regular expressions, as they often contain backslashes that need to be treated as literal characters.

This especially useful in regex patterns, where backslashes are commonly used to escape special characters or to denote character classes. By using a raw string, you can avoid the need to double backslashes in your regex patterns, making them easier to read and write.

Aside from that, we can also check if a string matches a specific pattern using the `re.match()` function, which checks for a match only at the beginning of the string.

For we want to check if the string starts with "A" and ends with ".", we can use the following pattern:
`r'^A.*\.$'`

```python
pattern = r'^A.*\.$'

items = [
    "An apple a day keeps the doctor away.",
    "A quick brown fox jumps over the lazy dog.",
    "The cat is on the roof.",
    "While the sun is shining, we will go to the park.",
]

for item in items:
    if re.match(pattern, item):
        print(f"[O] '{item}' matches the pattern.")
    else:
        print(f"[X] '{item}' does not match the pattern.")
```

## Data Validation

Data validation is the process of ensuring that input data is in the correct format and meets certain criteria before it is processed or stored. Regular expressions can be used for data validation by defining patterns that the input data must match.

For example, given name and password input, we want to check these following criteria:
- At least 8 characters long
- At least 1 digit

With those two criteria, we can use the following regular expression pattern:
`r'^(?=.*\d).{8,}$'`

- `^` asserts the start of the string.
- `(?=.*\d)` is a positive lookahead that asserts that there is at least one digit in the string. The `.*` allows for any characters to appear before the digit.
- `.{8,}` matches any character (except for line terminators) between 8 and unlimited times.
- `$` asserts the end of the string.


```python
pattern = r'^(?=.*\d).{8,}$'

passwords = [
    "password123",
    "12345678",
    "pass",
    "password",
    "pass1234",
]

for password in passwords:
    if re.match(pattern, password):
        print(f"[O] '{password}' is a valid password.")
    else:
        print(f"[X] '{password}' is not a valid password.")
```

## Data Extraction

Regex can also be used for data extraction, which is the process of extracting specific information from a larger text. The main benefit of using regex for data extraction is that it allows you to define complex patterns to match and extract specific pieces of information.

In this example we want to find all the entity in the text. In this case we want to find subset of the text that begin with uppercase letter and span across multiple words.

We can use the following regular expression pattern:
`r'\b(?:[A-Z]\w*)(?:\s+[A-Z]\w*)+\b'`

- `\b` asserts a word boundary, ensuring that we start matching at the beginning of a word.
- `(?:[A-Z]\w*)` matches a word that starts with an uppercase letter followed by zero or more word characters. The `(?:...)` syntax is a non-capturing group, which means that it groups the pattern without creating a separate capturing group in the result.
- `(?:\s+[A-Z]\w*)+` matches one or more occurrences of a space followed by a word that starts with an uppercase letter. The `+` quantifier indicates that this pattern can occur one or more times, allowing us to match multi-word entities.
- `\b` asserts another word boundary at the end of the match, ensuring that we end matching at the end of a word.

The `?:` is python specific syntax for non-capturing group, which means that the group will not be included in the result of the match. This is useful when you want to group parts of the pattern together without creating a separate capturing group for them.

Capturing groups are created using parentheses `()`, and they allow you to extract specific parts of the matched string.

```python
pattern = r'\b(?:[A-Z]\w*)(?:\s+[A-Z]\w*)+\b'

entities = re.findall(pattern, news)
entities
```

Other than `re.findall()`, we can also use `re.finditer()` to find all occurrences of the pattern in the string. The difference is that `re.finditer()` returns an iterator that produces match objects, which contain information about the match, such as the start and end positions of the match in the string. This can be useful if you want to perform additional processing on the matches or if you want to access specific groups within the matches.

```python
pattern = r'\b(?:[A-Z]\w*)(?:\s+[A-Z]\w*)+\b'

for entity in re.finditer(pattern, news):
    print(f"Found '{entity.group()}' at position {entity.start()}-{entity.end()}")
```

The pattern so far might be a bit cryptic, but we also can use simple pattern to extract specific information.

Let's say we want to get all the word 'nature', 'wildlife', and 'environment' in the news. We can use the following pattern:
`r'\b(?:nature|wildlife|environment)\b'`

- `\b` asserts a word boundary, ensuring that we match only whole words.
- `(?:nature|wildlife|environment)` is a non-capturing group that matches any of the specified words: "nature", "wildlife", or "environment". The `|` character is used to separate the alternatives within the group.
- `\b` asserts another word boundary at the end of the match, ensuring that we match only whole words.

Adding the flag `re.IGNORECASE` allows the regex to match the specified words regardless of their case (e.g., "Nature", "NATURE", "nature" will all be matched).

```python
pattern = r'\b(?:nature|wildlife|environment)\b'

for match in re.finditer(pattern, news, flags=re.IGNORECASE):
    print(f"Found '{match.group()}' at position {match.start()}-{match.end()}")
```

With the power of Python, we can specify the pattern in more dynamic way. For example the list of words that we want to match can be generated dynamically, and then we can create the regex pattern using string formatting or concatenation.

```python
# The list of words to match can be generated dynamically,
# for example from another source of data.
words_to_match = ["nature", "wildlife", "environment"]

# The join will result in 'nature|wildlife|environment'
pattern = r'\b(?:' + '|'.join(words_to_match) + r')\b'

for match in re.finditer(pattern, news, flags=re.IGNORECASE):
    print(f"Found '{match.group()}' at position {match.start()}-{match.end()}")
```

## More Complex Applications

With pattern matching, we can also manipulate the text in more complex way. For example, we can use regex to replace certain patterns in the text with new values.

### Replacing Patterns

Lets use the previous entities pattern to replace all the entity in the news with a placeholder string "[ENTITY]". We can use the `re.sub()` function for this purpose, which allows us to specify a replacement string for the matched pattern.

```python
pattern = r'\b(?:[A-Z]\w*)(?:\s+[A-Z]\w*)+\b'

replaced_news = re.sub(pattern, "[ENTITY]", news)

print("Before replacement:")
print(news)

print("After replacement:")
print(replaced_news)
```

Or we can also add some additional information to the matched pattern, for example for entity found we want to 'bold' the entity in the text by wrapping it with bracket `[` `]`. We can use a capturing group to capture the matched entity and then reference it in the replacement string.

```python
# We add a capturing group around the entity pattern to capture the matched entity
pattern = r'\b((?:[A-Z]\w*)(?:\s+[A-Z]\w*)+)\b'

# \1 refer to the first capturing group, which is the matched entity
replaced_news = re.sub(pattern, r'[\1]', news)

print("Before replacement:")
print(news)

print("After replacement:")
print(replaced_news)
```

### Filtering Text

In above news, lets assume each sentence is separated by a newline character `\n`. We can filter the sentences that contain some keywords. This is additional step after the above pattern matching.

First we split the news into sentences using the newline character as a delimiter, and then we can use regex for filtering the sentences that contain the keywords. We can use the same regex pattern for matching the keywords in each sentence.

For this example we want only sentences that must contain the word 'classroom' and pattern 'natur*'. So if only one of the keyword is found in the sentence, we will not include that sentence in the result. We can use the following pattern:
`r'^(?=.*\bclassroom\b)(?=.*\bnatur.*\b).*$'`

- `^` asserts the start of the string.
- `(?=.*\bclassroom\b)` is a positive lookahead that asserts that there is at least one occurrence of the word "classroom" in the string. The `.*` allows for any characters to appear before the word "classroom".
- `(?=.*\bnatur.*\b)` is another positive lookahead that asserts that there is at least one occurrence of a word that starts with "nature" in the string. The `.*` allows for any characters to appear before and after the word that starts with "nature".
- `.*$` matches any characters until the end of the string.

```python
pattern = r'^(?=.*\bclassroom\b)(?=.*\bnatur.*\b).*$'

sentences = news.split('\n')
filtered_sentences = []
for sentence in sentences:
    if re.match(pattern, sentence, flags=re.IGNORECASE):
        filtered_sentences.append(sentence)

for sentence in filtered_sentences:
    print(sentence)
```
