## Variables and Data Types

Python is a dynamically typed language, which is a fancy way of saying you do not need to tell the computer what kind of data you are storing; it figures it out for you. 

In this first cell, we explore the most common `primitive` types: strings for text, integers and floats for numbers, and booleans for True/False logic.

You should notice that we don't need semicolons or special keywords to declare these variables. We can also use the `type()` function to peek under the hood and see how Python is categorizing our data.

```python
# 1. Variable Assignment
name = "Alice"  # String
age = 30  # Integer
height = 5.8  # Float
is_learning = True  # Boolean

# 2. Printing and Type Checking
print(f"Hello, {name}!")
print("Name type: ", type(name))
print("Age type: ", type(age))
print("Height type: ", type(height))
print("Condition type: ", type(is_learning))

# 3. Dynamic Typing (Watch how 'age' changes)
age = "Thirty"
print("New age value: ", age)
print("New age type: ", type(age))
```

## Data Collections

Python has several built-in collections, which are data structures that can hold multiple items. The most common ones are lists, tuples, sets, and dictionaries. Each of these collections has its own unique properties and use cases.

1. **Lists** are ordered, mutable collections that can hold a variety of data types. They are defined using square brackets `[]`.
2. **Tuples** are ordered, immutable collections that can also hold a variety of data types. They are defined using parentheses `()`.
3. **Sets** are unordered collections of unique items. They are defined using curly braces `{}`.
4. **Dictionaries** are unordered collections of key-value pairs. They are also defined using curly braces `{}`, but they use a colon `:` to separate keys and values.

```python
# 1. Lists (Ordered)
programming_languages = ["Python", "Java", "C++", "JavaScript"]
print("First language:", programming_languages[0])
programming_languages.append("Rust")  # Adding to the list

# 2. Tuples (Ordered, Immutable)
coordinates = (10.0, 20.0)
print("Coordinates:", coordinates)

# This will raise an error because tuples are immutable; uncomment to see the error
# coordinates[0] = 15.0

# 3. Sets (Unordered, Unique)
unique_numbers = {1, 2, 3, 4, 5}
print("Unique numbers:", unique_numbers)
unique_numbers.add(3)  # This will not change the set because 3 is already in it
print("Unique numbers after adding 3 again:", unique_numbers)

# 4. Dictionaries (Unordered, Key-Value Pairs)
user_profile = {"username": "coder_123", "level": 5, "skills": programming_languages}

print(f"User {user_profile['username']} is level {user_profile['level']}")
print("The user knows:", user_profile["skills"][-1])  # Accessing the last item
```

**Small caveat**: In Python 3.7 and later, dictionaries maintain insertion order, but they are still considered unordered collections in terms of their conceptual design.

## Control Flow and Indentation

This is the most critical part of Python's syntax: the use of whitespace. 

Unlike other languages that use curly braces, Python uses indentation to define blocks of code. If you miss a colon or fail to indent properly, the code will fail. We use `if-elif-else` statements to make decisions and `for` loops to repeat actions.

In this example, we iterate through a list to see if a specific condition is met, demonstrating how logic and data collections work together seamlessly.

```python
# 1. Conditional Logic (If-Elif-Else Statements)
temperature = 25

if temperature > 30:
    print("It's a hot day.")
elif temperature >= 20:
    print("It's a pleasant day.")
else:
    print("It's a bit chilly.")

# elif is optional
# you can have as many elif blocks as needed, or none at all.
if temperature > 30:
    print("It's a hot day.")
else:
    print("It's not a hot day.")

# 2. Loops (Automating Repetition)
numbers = [1, 2, 3, 4, 5]
squared_numbers = []

for num in numbers:
    result = num ** 2 # Exponentiation
    squared_numbers.append(result)

print("Original:", numbers)
print("Squared:", squared_numbers)
```

## Functions and Modules

To avoid writing the same code over and over, we wrap our logic into Functions using the `def` keyword. Functions take inputs (parameters) and can give back a result (return). We also introduce Modules, which allow us to import code written by others.

Python's standard library is massive; by using the `import` statement, you can instantly access tools for math, dates, or even generating random numbers without having to build them from scratch.

```python
import math # Importing the math module

# 1. Defining a Function
def calculate_circle_area(radius):
    """Calculates the area of a circle given its radius."""
    area = math.pi * (radius ** 2)
    return round(area, 2)

# 2. Calling the Function
my_radius = 5
print(f"The area of a circle with radius {my_radius} is: {calculate_circle_area(my_radius)}")
```

Function can also have default values for parameters, making them flexible and reusable in different contexts.

To apply this, just specify the parameter name and assign it a default value in the function definition. This way, if you call the function without providing that parameter, it will use the default value instead.

```python
def hello(name="World"):
    return f"Hello, {name}!"

print(hello())  # Uses default value
print(hello("Alice"))  # Overrides default value
```

Another concept of functions is 'methods', which are functions that are associated with objects. For example, strings have methods like `.upper()` to convert text to uppercase, and lists have methods like `.append()` to add items to the end of the list.

```python
# Methods of Strings
name = "Alice"
print(name.upper())  # Output: ALICE
print(name.lower())  # Output: alice
print(name.replace("A", "E"))  # Output: Elice

# Methods of Lists
fruits = ["apple", "banana", "cherry"]
fruits.append("date")  # Adding an item to the end of the list
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date']
fruits.remove("banana")  # Removing an item from the list
print(fruits)  # Output: ['apple', 'cherry', 'date']
```

## More on Data Collections

### List Comprehensions

Some advanced concept of data collections include list comprehensions, generator expressions, and the `collections` module which provides specialized container datatypes like `defaultdict`, `OrderedDict`, and `Counter`.

```python
# list comprehension
squared_numbers = [x**2 for x in range(10)]
print("Squared numbers from 0 to 9:", squared_numbers)

# generator expression
# this will lazy-evaluate the squares of numbers from 0 to 9
# meaning it will generate each value on-the-fly as you iterate over it
# rather than storing them all in memory at once.
squared_gen = (x**2 for x in range(10))
print("Squared numbers from generator expression: ", list(squared_gen))

from collections import Counter

word_count = Counter([1, 2, 2, 3, 3, 3])
print("Word count: ", word_count)
```

### Collection Unpacking

Another quality of life feature of Python is 'collection unpacking', which allows you to assign values from a collection to multiple variables in a single line. For example:

```python
# Unpacking the tuple into variables x and y
coordinates = (10.0, 20.0)
x, y = coordinates
print(f"X: {x}, Y: {y}")

# Unpacking the list into variables
brothers = ["John", "Mike", "David"]
first_brother, second_brother, third_brother = brothers
print("The third brother is: ", third_brother)

# Unpacking with dictionaries
profile = {"name": "Alice", "age": 30, "city": "New York"}
for key, value in profile.items():
    print(f"{key}: {value}")
```

```python
# Unpacking should match the number of items in the collection
# If there are more variables than items, it will raise a ValueError.
a, b, c = [1, 2]
```

### Strings as Collections

From the defintion, string in python is collection of characters.

Therefore we can use any methods that we can use on string, we can also use on list, tuple, set, and dictionary. For example, we can use `len()` function to get the length of a collection, or we can use `in` keyword to check if an item is in a collection.

But the difference is that string is immutable, while list, tuple, set, and dictionary are mutable. This means that we cannot change the characters in a string, but we can change the items in a list, tuple, set, and dictionary. In this sense string is more like a tuple, because both of them are immutable.

```python
name = "Alice"

# Iterating through each character in the string 'name'
for char in name:
    print(char)

print("Length of name:", len(name))
print("Is 'A' in name?", 'A' in name)
print("Is 'a' in name?", 'a' in name)
print("The second character in name is:", name[1])
```

```python
# Assinment of character in string will raise an error because string is immutable
name[0] = 'E'
```

Slicing is another powerful feature of Python that allows you to access a subset of a collection. For example, you can slice a list to get a portion of its elements, or you can slice a string to get a substring. Slicing uses the syntax `collection[start:stop:step]`, where `start` is the index to begin slicing, `stop` is the index to end slicing (exclusive), and `step` is the interval between indices.

```python
# range() is a built-in function that generates a sequence of numbers
numbers = list(range(10))  # This will create a list of numbers from 0 to 9
print("Numbers from 0 to 9:", numbers)

# collection[start:stop:step]; start with 1, stop at the end, and step by 2 to get odd numbers
odd_numbers = numbers[1::2]
print("Odd numbers:", odd_numbers)

# -1 is used to reverse the order of the collection
reversed_numbers = numbers[::-1]
print("Reversed numbers:", reversed_numbers)
```

```python
# -1 also can be used to access the index from the end of the collection
last_number = numbers[-1]
print("Last number:", last_number)

second_last_number = numbers[-2]
print("Second last number:", second_last_number)
```
